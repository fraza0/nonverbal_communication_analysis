## Composed Background Images

```python
pc1_bg = 'DATASET_DEP/BACKGROUND/pc1_bg.png'
pc2_bg1 = 'DATASET_DEP/BACKGROUND/pc2_bg_1.png'
pc2_bg2 = 'DATASET_DEP/BACKGROUND/pc2_bg_2.png'
pc3_bg1 = 'DATASET_DEP/BACKGROUND/pc3_bg_1.png'
pc3_bg2 = 'DATASET_DEP/BACKGROUND/pc3_bg_2.png'

img_pc1_bg = cv2.imread(pc1_bg)
img_pc2_bg1 = cv2.imread(pc2_bg1)
img_pc2_bg2 = cv2.imread(pc2_bg2)
img_pc3_bg1 = cv2.imread(pc3_bg1)
img_pc3_bg2 = cv2.imread(pc3_bg2)
```

```python
height, width, depth = img_pc2_bg1.shape
width_cutoff = width // 2
img_pc2_bg = np.concatenate((img_pc2_bg2[:, :width_cutoff], img_pc2_bg1[:, width_cutoff:]), axis=1)
#cv2.imwrite('pc2_bg.png', img_pc2_bg)
plt.imshow(img_pc2_bg)
```

```python
height, width, depth = img_pc3_bg1.shape
width_cutoff = width // 2
img_pc3_bg = np.concatenate((img_pc3_bg1[:, :width_cutoff], img_pc3_bg2[:, width_cutoff:]), axis=1)
#cv2.imwrite('pc3_bg.png', img_pc3_bg)
plt.imshow(img_pc3_bg)
```

## Background Subtraction

### Direct / Raw approach

```python
test_frame_pc1_path = ''
test_frame_pc2_path = ''
test_frame_pc3_path = 'DATASET_DEP/SYNC/3CLC9VWRSAMPLE/task_1/frames/cut001.png'

test_frame_pc1 = cv2.imread(test_frame_pc1_path)
test_frame_pc2 = cv2.imread(test_frame_pc2_path)
test_frame_pc3 = cv2.imread(test_frame_pc3_path)
```

```python
direct_bg_sub_pc3 = cv2.absdiff(test_frame_pc3, img_pc3_bg)
plt.imshow(direct_bg_sub_pc3)
```

### Smoothing

```python
smooth_test_pc3 = cv2.blur(test_frame_pc3,(5,5))
smooth_bg_pc3 = cv2.blur(img_pc3_bg, (5,5))
smooth_bg_sub_pc3 = cv2.absdiff(smooth_test_pc3, smooth_bg_pc3)
plt.imshow(smooth_bg_sub_pc3)
```
